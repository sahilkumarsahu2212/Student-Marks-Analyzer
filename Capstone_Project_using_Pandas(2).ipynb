```python
#Student Marks Analyzer
print("Goal:   Analyze marks of students in one subject using Series")
print("======================================================================")
print("Problem Statement:")
print("-"*70)
print("Given marks of Students,find:")
print("\t\t1.Class Average and Topper")
print("\t\t2.Replace Missing Marks with the mean")
print("\t\t3.Show top 3 Students")
print("-"*70)
import pandas as pd
import numpy as np
marks=pd.Series([85,90,np.nan,75,88,92,np.nan,None],index=['Sahil','Deepali','Raja','Liza','Rasmita','Rama','Subha','Jigu'])
print("Original Marks:\n",marks)
#1.Class Average and Toppers
print()
print("Class Average:",marks.mean())
print("Class Topper:",marks.idxmax(),"=>",marks.max())
print()
#2.Replace Missing Marks with the mean
print("Replace Missing Marks with the mean:")
marks=marks.fillna(marks.mean())
print(marks)
print()
#3.Show Top 3 Students
print("Top 3 Students are:")
marks.sort_values(ascending=False,inplace=True)
print(marks.head(3))
```
